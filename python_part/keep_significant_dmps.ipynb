### AVPC DMPs

```python
# DATA_PATH = "./data/avpc/all_DMPs/avpc_vs_normal_DMPs.csv"
# SIG_DMP_PATH = "./data/avpc/significant_DMPs/avpc_vs_normal_sig_DMPs.csv"

# DATA_PATH = "./data/avpc/all_DMPs/avpc_vs_indolent_DMPs.csv"
# SIG_DMP_PATH = "./data/avpc/significant_DMPs/avpc_vs_indolent_sig_DMPs.csv"

DATA_PATH = "./data/avpc/all_DMPs/avpc_vs_high_grade_DMPs.csv"
SIG_DMP_PATH = "./data/avpc/significant_DMPs/avpc_vs_high_grade_sig_DMPs.csv"
```

```python
# Load data
df = pd.read_csv(DATA_PATH)
df['neg_log10_pval'] = -np.log10(df['P.Value'])
df
```

```python
# Define differential expression status
df['diffexpressed'] = 'NO'
df.loc[(df['logFC'] < -1) & (df['P.Value'] < 0.001), 'diffexpressed'] = 'HYPO'
df.loc[(df['logFC'] > 1) & (df['P.Value'] < 0.001), 'diffexpressed'] = 'HYPER'
```

```python
# Filter all downregulated and upregulated
downregulated = df[df['diffexpressed'] == 'HYPO'].sort_values(by=['neg_log10_pval', 'logFC'], ascending=[False, True])
upregulated = df[df['diffexpressed'] == 'HYPER'].sort_values(by=['neg_log10_pval', 'logFC'], ascending=[False, False])
```

```python
# Combine all downregulated and upregulated gene names
# Select only the 'gene_name' and 'diffexpressed' columns
downregulated_filtered = downregulated[['Name', 'logFC','P.Value', 'diffexpressed']]
upregulated_filtered = upregulated[['Name', 'logFC','P.Value', 'diffexpressed']]

# Concatenate the downregulated and upregulated data
result_df = pd.concat([downregulated_filtered, upregulated_filtered], ignore_index=True)

# Display the resulting DataFrame
result_df
```

```python
# Save the DataFrame to a CSV file without the index
result_df.to_csv(SIG_DMP_PATH, index=False)
```
