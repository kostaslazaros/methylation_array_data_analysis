### AVPC DMPs

```python
SIG_DMP_PATH_01 = "./data/avpc/significant_DMPs/avpc_vs_normal_sig_DMPs.csv"
SIG_DMP_PATH_02 = "./data/avpc/significant_DMPs/avpc_vs_indolent_sig_DMPs.csv"
SIG_DMP_PATH_03 = "./data/avpc/significant_DMPs/avpc_vs_high_grade_sig_DMPs.csv"
```

### Intersection Paths

```python
# INTERSECTION_PATH = "./data/intersection_lists/normal_intersection.xlsx"

# INTERSECTION_PATH = "./data/intersection_lists/indolent_intersection.xlsx"

# INTERSECTION_PATH = "./data/intersection_lists/high_grade_intersection.xlsx"

INTERSECTION_PATH = "./data/intersection_lists/avpc_intersection.xlsx"
```

```python
COLS2KEEP = ["Name", "logFC", "P.Value"]
```

```python
df1 = pd.read_csv(SIG_DMP_PATH_01)
df2 = pd.read_csv(SIG_DMP_PATH_02)
df3 = pd.read_csv(SIG_DMP_PATH_03)
```

```python
df1 = df1[COLS2KEEP]
df2 = df2[COLS2KEEP]
df3 = df3[COLS2KEEP]
```

```python
# Merging three dataframes on 'Name' column with unique suffixes for each
merged_df = df1.merge(df2, on="Name", suffixes=("_1", "_2")).merge(
    df3.add_suffix("_3").rename(columns={"Name_3": "Name"}), on="Name"
)
merged_df
```

```python
# Calculate mean for logFC and P.Value across the three dataframes
merged_df["logFC_median"] = merged_df[["logFC_1", "logFC_2", "logFC_3"]].median(axis=1)
merged_df["P.Value_median"] = merged_df[["P.Value_1", "P.Value_2", "P.Value_3"]].median(axis=1)
merged_df
```

```python
# Create final_df as a copy to avoid SettingWithCopyWarning
final_df = merged_df[["Name", "logFC_median", "P.Value_median"]].copy()

# Add the 'diff_meth' column based on the 'logFC_mean' values
final_df["diff_meth"] = np.where(final_df["logFC_median"] < 0, "HYPO", "HYPER")

final_df
```

```python
# Save the resulting DataFrame to a CSV file without the index
final_df.to_excel(INTERSECTION_PATH, index=False)
```
