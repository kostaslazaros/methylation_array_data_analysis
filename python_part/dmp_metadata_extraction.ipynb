```python
import pandas as pd
```

```python
sigdmp_df = pd.read_excel("./data/intersection_lists/merged_unique_DMPs.xlsx")
sigdmp_df
```

```python
epicv2_meta = pd.read_excel("./data/EPICv2_source_data.xlsx", index_col=0)
epicv2_meta
```

```python
significant_dmps = sigdmp_df['significant DMPs'].values
```

```python
# Filter df2 to keep only rows where the index is in significant_dmps
epicv2_meta = epicv2_meta[epicv2_meta.index.isin(significant_dmps)]
epicv2_meta
```

```python
non_nan_count = epicv2_meta['Methyl450_Loci'].count()
non_nan_count
```

```python
# Replace all NaN values with the word "Unknown"
epicv2_meta = epicv2_meta.fillna("Unknown")
epicv2_meta
```

```python
epicv2_meta = epicv2_meta.replace("Unknown", "N/A")
epicv2_meta
```

```python
epicv2_meta.to_excel("./data/significant_dmps_epicv2_metadata.xlsx", index=False)
```
