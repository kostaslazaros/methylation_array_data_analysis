```python
import pandas as pd
```

```python
df = pd.read_excel("./data/significant_dmps_epicv2_metadata.xlsx")
df
```

```python
# Select the two columns
df_filtered = df[['ID_source', 'Methyl450_Loci', 'EPICv1_Loci']]

# Drop rows with NaN values in these columns
df_filtered = df_filtered.dropna(subset=['ID_source', 'Methyl450_Loci', 'EPICv1_Loci'])
df_filtered
```

```python
df_filtered.to_excel("./data/significant_dmps_in_all_array_designs.xlsx", index=False)
```

```python
# Select the two columns
df_nan_filtered = df[['ID_source', 'Methyl450_Loci', 'EPICv1_Loci']]
# Filter rows where both 'Column1' and 'Column2' are NaN
df_nan_filtered = df_nan_filtered[df_nan_filtered[['Methyl450_Loci', 'EPICv1_Loci']].isna().all(axis=1)]
df_nan_filtered
```

```python
df_nan_filtered.to_excel("./data/significant_dmps_epicv2_exclusive.xlsx", index=False)
```
